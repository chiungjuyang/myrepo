## 串列 (list)
### - append

```python
a = []
b = [1,2.0,"3",[4],(5)]
print(b[1])
print(b[3])
print(b)
```

```python
list1 = [1,2,3,4]
print(list1)
list2 = ["a","b","c"]
list1.append(list2)
print(list1)
```

### - extend

```python
list1 = [1,2,3,4]
print(list1)
list2 = ["a","b","c"]
list1.extend(list2)
print(list1)
list1.extend("test")
print(list1)
```

### - insert

```python
list1 = [1,2,3,4]
print(list1)
list1.insert(2,"insert")
print(list1)
list1.insert(7,"insert2")
print(list1)
list1.insert(6,"insert3")
print(list1)
```

### - remove

```python
list1 = [1,"a",(12.4),[True],range(10)]
print(list1)
list1.remove(1)
print(list1)
list1.remove("a")
print(list1)
list1.remove(range(10))
print(list1)
list1.remove((12.4))
print(list1)
list1.remove([True])
print(list1)
```

```python
list1 = [1,"a",(12.4),[True],range(10)]
print(list1)
list1.pop(1)
print(list1)
list1.pop()
print(list1)
```

```python
L = [1,3,5,7,9]
#L.remove(1)
L.pop(1)
L
```

### - index

```python
list1 = [1, 2, 3, 5, 1, 47, 65, 45, 14, 2, 4, 5, 7, 4, 32, 1, 8, 4, 6]
print(list1)
print(list1.index(2))
print(list1.index(2,7))
print(list1.index(4,5,15))
```

### - count
### - sort
### - reverse
### - copy
### - pop

```python
stu = ["john","marry","ben","adam"]
print(stu)
stu.append("ramsay")
print(stu)
stu2 = ["dion","carl","john"]
stu.extend(stu2)
print(stu)
stu.extend("test")
print(stu)
stu.remove("john")
print(stu)
print(stu.pop(2))
print(stu)
```

```python
stu.insert(0,"dion")
print(stu)
print(stu.count("dion"))
stu.sort()
print(stu)
stu.reverse()
print(stu)
stu3 = stu
stu4 = stu.copy()
print(stu == stu3)
print(stu is stu3)
print(stu == stu4)
print(stu is stu4)
```

```python
L = [3,6,1,2,9]
L.sort()
print(L)
L.reverse()
print(L)
```

```python
L = [3,6,1,2,9]
L.sort(reverse=True)
print(L)
```
